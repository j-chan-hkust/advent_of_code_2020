seems like this question is just a binary number question in disguise.
Obviously, F == 0, B == 1,
and similarly, R == 1, L == 0

```python
def convert_to_binary(input_str):
    output = ""
    for char in input_str:
        if char == "F" or char == "L":
            output += '0'
        else:
            output += '1'
    return output
```
