So evidently, trying to change the second last item doesn't solve the problem. In that case, the "problematic" instruction must be in the history of previous instructions. So, what we should be doing is keeping a log of previously run instructions, and checking which one causes the code to reach the end

```python
inputs_and_status = [[item, None] for item in inputs]
inputs_and_status

prev_states = []
current_state = 0 #we start from operation 0
current_acc = 0
while inputs_and_status[current_state][1] == None:
    if inputs_and_status[current_state][0][0:3] == "nop":
        inputs_and_status[current_state][1] = current_acc
        prev_states.append(current_state)
        current_state += 1
    elif inputs_and_status[current_state][0][0:3] == "jmp":
        inputs_and_status[current_state][1] = current_acc
        prev_states.append(current_state)
        current_state += int(inputs_and_status[current_state][0][4:])
    elif inputs_and_status[current_state][0][0:3] == "acc":
        current_acc += int(inputs_and_status[current_state][0][4:])
        inputs_and_status[current_state][1] = current_acc
        prev_states.append(current_state)
        current_state += 1

print(prev_states)
```

```python
for state in prev_states:
    temp_input = inputs.copy()
    if temp_input[state][0:3] == "jmp":
        temp_input[state] = "nop" + temp_input[state][3:]
    elif temp_input[state][0:3] == "nop":
        temp_input[state] = "jmp" + temp_input[state][3:]
    else:
        continue
    
    inputs_and_status = [[item, None] for item in temp_input]
    inputs_and_status

    current_state = 0 #we start from operation 0
    current_acc = 0
    while inputs_and_status[current_state][1] == None:
        if inputs_and_status[current_state][0][0:3] == "nop":
            inputs_and_status[current_state][1] = current_acc
            current_state += 1
        elif inputs_and_status[current_state][0][0:3] == "jmp":
            inputs_and_status[current_state][1] = current_acc
            current_state += int(inputs_and_status[current_state][0][4:])
        elif inputs_and_status[current_state][0][0:3] == "acc":
            current_acc += int(inputs_and_status[current_state][0][4:])
            inputs_and_status[current_state][1] = current_acc
            current_state += 1
        if len(inputs_and_status) == current_state:
            print(current_acc)
            print(current_state)
```
